```python
data = df[['GENDER','Car_Owner','Propert_Owner','Annual_income','Marital_status','EDUCATION','label']]
data
```

```python
df['EDUCATION'].unique()
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
data['Car_Owner'] = labelencoder.fit_transform(data['Car_Owner'])
data['GENDER'] = labelencoder.fit_transform(data['GENDER'])
data['Propert_Owner'] = labelencoder.fit_transform(data['Propert_Owner'])
data['Marital_status'] = labelencoder.fit_transform(data['Marital_status'])
data['EDUCATION'] = labelencoder.fit_transform(data['EDUCATION'])
```

```python
data
```

```python
X = data.drop('label', axis='columns')
X
```

```python
y =data['label']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print('Accuracy score is {}'.format(accuracy_score(y_test, predictions)))
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
scaled_X_train,y_train = sm.fit_resample(scaled_X_train, y_train)
```

```python
y_test.unique()
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train, y_train)
```

```python
log_predictions = log_model.predict(scaled_X_test)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {'n_neighbors' : [3,5,7,9,11],
               'weights' : ['uniform','distance'],
               'algorithm' : ['auto', 'kd_tree']}

gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=2)               
```

```python
gridkn.fit(scaled_X_train, y_train)
```

```python
gridkn.best_params_
```

```python
knpreds = gridkn.predict(scaled_X_test)
```

```python
modelperformance(knpreds)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid_svr = {"C":[0.01,0.1,0.5,1],
                  'kernel': ['linear', 'rbf','poly']}

gridsvr = GridSearchCV(svm,param_grid_svr)                 
```

```python
gridsvr.fit(scaled_X_train,y_train)
```

```python
pred_svc = gridsvr.predict(scaled_X_test)
modelperformance(pred_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
param_grid = {
    'criterion' :['gini','entropy'],
    'splitter': ['best','random'],
    'max_depth' : [None, 10,20,30,40,50],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4]
}

grid_search_tree = GridSearchCV(estimator=DecisionTreeClassifier(),param_grid=param_grid)
```

```python
grid_search_tree.fit(scaled_X_train, y_train)
```

```python
preds = grid_search_tree.predict(scaled_X_test)
modelperformance(preds)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
n_estimators = [32,64,128,256]
max_features = [2,3,4]
bootstrap = [True, False]
oob_score = [True, False]
```

```python
param_grid_rfc = {
    'n_estimators': n_estimators,
    'max_features' : max_features,
    'bootstrap' : bootstrap,
    'oob_score': oob_score
}
```

```python
grid_rfc = GridSearchCV(rfc,param_grid_rfc)
```

```python
grid_rfc.fit(scaled_X_train, y_train)
```

```python
grid_rfc.best_params_
```

```python
pred_rfc = grid_rfc.predict(scaled_X_test)
modelperformance(pred_rfc)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

base_tree = DecisionTreeClassifier(max_depth=1,class_weight='balanced')
ada_classifier = AdaBoostClassifier(estimator=base_tree)
ada_param_grid = {
    'n_estimators':[50,100,200],
    'learning_rate':[0.01,0.1,1,10]
}

ada_grid= GridSearchCV(ada_classifier, ada_param_grid)
```

```python
ada_grid.fit(scaled_X_train, y_train)
```

```python
ada_pred = ada_grid.predict(scaled_X_test)
modelperformance(ada_pred)
```
